# RandomForest

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('ai_dev_productivity.csv')
```

```python
df.head(2)
```

```python
# df.shape
```

```python
# df['task_success'].value_counts()
```

```python
# df.isnull().sum()
```

```python
x=df.drop(columns = ['task_success'])
y = df['task_success']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_train, y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

# RandomForestRegressor

```python
df = pd.read_csv('insurance.csv')
```

```python
df.head(2)
```

```python
df = pd.get_dummies(df, columns = ['sex', 'smoker', 'region'])
```

```python
df.head(2)
```

```python
x=df.drop(columns=['charges'])
y = df['charges']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor()
```

```python
rf.fit(x_train, y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

# R.F. Classifier

```python
df = pd.read_csv('attrition.csv')
```

```python
df.head(2)
```

```python
# df.dtypes
```

```python
df_obj = ['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'Over18', 'OverTime']
```

```python
df = pd.get_dummies(df, columns = df_obj)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
df['Attrition'] = lb.fit_transform(df['Attrition'])
```

```python
x = df.drop(columns = ['Attrition'])
y = df['Attrition']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_train, y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```
